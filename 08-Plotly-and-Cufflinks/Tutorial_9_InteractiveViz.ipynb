```python
# plotly image just use iplot(p;otly interactive images)
df.iplot()

```

### different types of Plots using IPLOTS

```python
df.iplot(kind='scatter',x='A',y='B')  #default ploty draws lines connecting the markes
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=20)

```

```python
df2
```

```python
#barplots
df2.iplot(kind='bar',x='Category',y='Values')  #x needs to be a category
```

```python
# bar plot cannot be used for all kind of data 
df
```

```python
df.iplot(kind='bar')  #bar for each of the value 
# but we can perform some sort of aggregation over them
```

```python
df.count().iplot(kind='bar')  # bar plot for each instance here just 100 for each
```

```python
df.sum().iplot(kind='bar')
```

```python
# boxPlot
```

```python
df.iplot(kind='box')  #for each column it will draw
```

```python
# 3-D surface plots
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3   #has three dimensions or three values
```

```python
# surface plot
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
# note plotly's colorscale don't exactly match with matplotlib colorscale so do check the official documentation for details
```

### Bubble plots


```python
# essentially like scatter plot but some difference
# used for happiness factors,population etc.
```

```python
df.iplot(kind='bubble',x='A',y='B',size='C')  #size of bubble is based on column C
```

```python
# useful in case of world GDP also
```
